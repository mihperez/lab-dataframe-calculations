```python
total_sales=orders.groupby(['CustomerID']).agg({"amount_spent": "sum"})

labels1=["others", "Preferred", "VIP"]
total_sales["Category"]=pd.qcut(total_sales['amount_spent'], [0, 0.75, 0.95, 1], labels=labels1)
total_sales.sort_values(by="amount_spent", ascending=False)






```

## Q2: How to identify which country has the most VIP Customers?

```python
total_sales_country = pd.merge(total_sales, orders, how='inner', on = 'CustomerID').reset_index()
condition=total_sales_country['Category']=='VIP'
total_sales_country[condition]

total_sales_country_agg=total_sales_country[condition].groupby(["Country"])["CustomerID"].nunique().reset_index(name='Unique_Customers').sort_values(by=('Unique_Customers'), ascending=False)
total_sales_country_agg


#total_sales_country_agg_VIP=total_sales_country_agg.query("Category=='VIP'")
#total_sales_country_agg_VIP.sort_values
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
total_sales_country = pd.merge(total_sales, orders, how='inner', on = 'CustomerID').reset_index()

condition= (total_sales_country['Category']!= 'others')
total_sales_country[condition]

total_sales_country_agg=total_sales_country[condition].groupby(["Country"])["CustomerID"].nunique().reset_index(name='Unique_Customers').sort_values(by=('Unique_Customers'), ascending=False)
total_sales_country_agg
```
